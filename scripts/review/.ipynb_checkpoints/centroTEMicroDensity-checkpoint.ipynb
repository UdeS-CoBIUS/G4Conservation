## get stats 

```python
densities = {}
dfDens = pd.DataFrame()
groups = dfMicro.groupby('Sp')
for spName, spDf in groups:
    groups2 = spDf.groupby('Repro')
    for repro, reproDf in groups2:
        groups3 = reproDf.groupby('ShuffType')
        for shuffType, shuffTypeDf in groups3:
            if spName not in densities:
                densities[spName] = {}
            if repro not in densities[spName]:
                densities[spName][repro] = {}
            nbpG4 = len(shuffTypeDf)
            dfTmp = dfNtMicro[dfNtMicro['Sp'] == spName]
            A = list(dfTmp[dfTmp['Type'] == 'micro sat G'].A)[0] + list(dfTmp[dfTmp['Type'] == 'micro sat no G'].A)[0]
            T = list(dfTmp[dfTmp['Type'] == 'micro sat G'].T)[0] + list(dfTmp[dfTmp['Type'] == 'micro sat no G'].T)[0]
            G = list(dfTmp[dfTmp['Type'] == 'micro sat G'].G)[0] + list(dfTmp[dfTmp['Type'] == 'micro sat no G'].G)[0]
            C = list(dfTmp[dfTmp['Type'] == 'micro sat G'].C)[0] + list(dfTmp[dfTmp['Type'] == 'micro sat no G'].C)[0]
            dens = nbpG4 / float(A+T+C+G) *1000
            densities[spName][repro][shuffType] = dens
            row = {'Sp': [spName], 'Repro': [repro], 'shuffType': [shuffType], 'Density': [dens], 'GC': float(C+G)/float(A+T+C+G) *100}
            dfDens = dfDens.append(pd.DataFrame(data=row))                
```

```python
dfHisto = pd.DataFrame()
for sp in list(set(dfDens.Sp)):
    dfTmp = dfDens[dfDens.shuffType == 'Wt']
    row = {'Sp': [sp], 'Mean': [list(dfTmp[dfTmp.Sp == sp].Density)[0]], 'Std': [0], 'Type': ['Wt'], 'GC': [list(dfTmp[dfTmp.Sp == sp].GC)[0]]}
    dfHisto = dfHisto.append(pd.DataFrame(data=row))
    dfTmp = dfDens[dfDens.shuffType == 'Mono']
    row = {'Sp': [sp], 'Mean': [dfTmp[dfTmp.Sp == sp].Density.mean()], 'Std': [dfTmp[dfTmp.Sp == sp].Density.std()], 'Type': ['Mono'], 'GC': [list(dfTmp[dfTmp.Sp == sp].GC)[0]]}
    dfHisto = dfHisto.append(pd.DataFrame(data=row))
    dfTmp = dfDens[dfDens.shuffType == 'Tri']
    row = {'Sp': [sp], 'Mean': [dfTmp[dfTmp.Sp == sp].Density.mean()], 'Std': [dfTmp[dfTmp.Sp == sp].Density.std()], 'Type': ['Tri'], 'GC': [list(dfTmp[dfTmp.Sp == sp].GC)[0]]}
    dfHisto = dfHisto.append(pd.DataFrame(data=row))
    
dfHisto.head()
```

## Graph

```python
dfDotPlot = pd.DataFrame()
groups = dfHisto.groupby('Sp')
for spName, spDf in groups:
    row = {'Sp': spName, 'Wt': [list(spDf[spDf.Type == 'Wt'].Mean)[0]], 'Mono': [list(spDf[spDf.Type == 'Mono'].Mean)[0]], 'GC': [list(spDf.GC)[0]], 'SpRealm': 'Eukaryotes'}
    dfDotPlot = dfDotPlot.append(pd.DataFrame(data=row))
dfDotPlot['DiffDens'] = dfDotPlot['Wt'] - dfDotPlot['Mono']

my_pl = {'Eukaryotes':"#D3772E", 'Archaea':"#2686a4", 'Bacteria':"#c33543"} #test1
# my_pl = {'Eukaryote':"#e9c662", 'Eukaryotes':"#e9c662", 'Archaea':"#73bca9", 'Bacteria':"#e74e46"} #test2
sns.lmplot( x="GC", y="DiffDens", data=dfDotPlot, fit_reg=False, hue='SpRealm',markers=["^"], legend=True, palette=my_pl, scatter_kws={"s": 100})
# plt.ylim([-0.5, +0.8])
plt.xlim([20, +75])
plt.title("pG4 Densities and GC content for Microsatellite sequences", loc='left')
# plt.savefig("FiguresArticle/dotPlotGeneMeanShuff.svg")
plt.savefig("/home/anais/Documents/Projet/G4Conservation/scripts/FiguresArticle/reviewCorrelation/dotPlotMicroSatMeanShuff.svg")
plt.show()
```

# Transposable elements

## import data

```python
path = '/home/anais/Documents/Projet/G4Conservation/'
dfTE = pd.DataFrame()

for repro in range(1,11):
    repro = str(repro)
    dfTmp = pd.read_csv(path+'reviewTRCentro/ManySpecies/Repro'+repro+'/pG4_Shuffle_Mono_Micro.csv', sep='\t')
    dfTmp['ShuffType'] = 'Mono'
    dfTmp['Repro'] = repro
    dfTE = dfTE.append(dfTmp)

    dfTmp = pd.read_csv(path+'reviewTRCentro/ManySpecies/Repro'+repro+'/pG4_Shuffle_Tri_Micro.csv', sep='\t')
    dfTmp['ShuffType'] = 'Tri'
    dfTmp['Repro'] = repro
    dfTE = dfTE.append(dfTmp)

dfTmp = pd.read_csv(path+'reviewTRCentro/ManySpecies/pG4_Shuffle_WT_Micro.csv', sep='\t')
dfTmp['ShuffType'] = 'Wt'
dfTmp['Repro'] = 0
dfTE = dfTE.append(dfTmp)
    
dfTE = dfTE[dfTE.pG4Start != 'pG4Start']
dfTE.head()
```

```python
def getDicoIdSp():
    dicoIdSp = {}
    with open(path+'reviewTRCentro/ManySpecies/IdSpeciesName.txt') as f:
        content = f.read()
        lines = content.split('\n')
        for l in lines:
            w = l.split('\t')
            if w[0] != '':
                if w[0][-1] == ' ':
                    dicoIdSp[w[0][:-1]] = w[1]
                else:
                    dicoIdSp[w[0]] = w[1]
    return dicoIdSp

def getSp(Description):
    dico = getDicoIdSp()
    return dico[Description]
            
dfTE['Sp'] = dfTE.Description.apply(getSp)
dfTE['Sp'] = np.where(dfTE['Sp'] == 'Escherichia coli', 'Ecol', dfTE['Sp'])
dfTE.head()
```

## get densities

```python
dico = {}
GC = {}
dicoId = getDicoIdSp()
filename = path+'reviewTRCentro/ManySpecies/Sequences_TE.fa'
fastaOrigin = SeqIO.parse(open(filename),'fasta')
for fasta in fastaOrigin:
    name, seq = fasta.id, str(fasta.seq)
    if name in dicoId:
        sp = dicoId[name]
    elif name+' ' in dicoId:
        sp = dicoId[name+' ']
    elif name+' sequence' in dicoId:
        sp = dicoId[name+' sequence']
    if sp not in dico:
        dico[sp] = 0
        GC[sp] = {'G': 0, 'C': 0, 'T': 0, 'A': 0, 'N': 0}
    dico[sp] += len(seq)
    GC[sp]['A'] += seq.count('A')
    GC[sp]['T'] += seq.count('T')
    GC[sp]['G'] += seq.count('G')
    GC[sp]['C'] += seq.count('C')
    GC[sp]['N'] += seq.count('N')
    GC[sp]['A'] += seq.count('a')
    GC[sp]['T'] += seq.count('t')
    GC[sp]['G'] += seq.count('g')
    GC[sp]['C'] += seq.count('c')
    GC[sp]['N'] += seq.count('n')

dico['Ecol'] += dico['Escherichia coli']
GC['Ecol']['A'] += GC['Escherichia coli']['A']
GC['Ecol']['T'] += GC['Escherichia coli']['T']
GC['Ecol']['G'] += GC['Escherichia coli']['G']
GC['Ecol']['C'] += GC['Escherichia coli']['C']
GC['Ecol']['N'] += GC['Escherichia coli']['N']

del dico['Escherichia coli']
del GC['Escherichia coli']
pprint(dico)
```

```python
densities = {}
dfDens = pd.DataFrame()
groups = dfTE.groupby('Sp')
for spName, spDf in groups:
    groups2 = spDf.groupby('Repro')
    for repro, reproDf in groups2:
        groups3 = reproDf.groupby('ShuffType')
        for shuffType, shuffTypeDf in groups3:
            if spName not in densities:
                densities[spName] = {}
            if repro not in densities[spName]:
                densities[spName][repro] = {}
            nbpG4 = len(shuffTypeDf)
            dfTmp = dfNtMicro[dfNtMicro['Sp'] == spName]
            dens = nbpG4 / float(dico[spName]) *1000
            densities[spName][repro][shuffType] = dens
            row = {'Sp': [spName], 'Repro': [repro], 'shuffType': [shuffType], 'Density': [dens], 'GC': float(GC[spName]['G']+GC[spName]['C'])/float(dico[spName]) *100}
            dfDens = dfDens.append(pd.DataFrame(data=row)) 
dfDens.head()
```
