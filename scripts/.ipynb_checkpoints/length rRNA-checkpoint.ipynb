```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
def appendDf(filename, sp):
    dftmp = pd.read_csv(filename, sep='\t')
    dftmp['Sp'] = sp
    return dftmp
```

```python
df = pd.DataFrame()
path = '/home/anais/Documents/Projet/G4Conservation/'

for path2, dirs, files in os.walk(path+'toDlLocal/'):
    for sp in dirs:
        if sp not in ['Figures', 'Fasta', 'CSVFile', 'SplitFile']:
            try :
                df = df.append(appendDf(path + 'toDlLocal/' + sp + '/pG4WT.csv', sp))
            except:
                print(sp + ' doesnt have any pseudogenes non coding transcripts.')
df['Length'] = df['pG4End'] - df['pG4Start']
df['Length'] = abs( df['Length'] )
df['LengthTr'] = df['locEnd'] - df['locStart']
print(list(set(df['Class'])))
df.head()
```

```python
# print(df[df.Class == 'ShortNC'].head())
dftmp = pd.DataFrame()
dftmp = dftmp.append(df[df.Biotype == 'rRNA'])
dftmp = dftmp[dftmp.Class == 'ShortNC']
dftmp = dftmp[['LengthTr', 'Sp']]
dftmp = dftmp.drop_duplicates(subset=None, keep='first', inplace=False)

sns.boxplot( x=dftmp['Sp'], y=dftmp['LengthTr'], color="#E64E46" )
plt.tight_layout()
```
